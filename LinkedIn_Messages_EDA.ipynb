```python
messages = pd.read_csv("messages.csv")
messages
```

```python
data_info = messages.info()
data_head = messages.head()

data_info, data_head
```

```python
# Replacing the variable name 'data' with 'messages' as requested

# Convert the 'DATE' column to datetime format for analysis
messages['DATE'] = pd.to_datetime(messages['DATE'])

# Extract the date and time components for further analysis
messages['DATE_ONLY'] = messages['DATE'].dt.date
messages['TIME_ONLY'] = messages['DATE'].dt.time

# Analyze the most frequent senders
top_senders = messages['FROM'].value_counts().head(10)

# Analyze the most frequent recipients
top_recipients = messages['TO'].value_counts().head(10)

# Analyze the most active dates (number of messages per day)
most_active_dates = messages['DATE_ONLY'].value_counts().head(10)

top_senders, top_recipients, most_active_dates
```

```python
top_senders.plot(kind='bar', color='skyblue')
plt.title('Top 10 Senders')
plt.xlabel('Sender')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45)
plt.show()
```

```python
top_recipients.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Recipients')
plt.xlabel('Recipient')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45)
plt.show()
```

```python
most_active_dates.plot(kind='bar', color='lightcoral')
plt.title('Top 10 Most Active Dates')
plt.xlabel('Date')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45)
plt.show()
```

```python
# Grouping by date to count the number of messages per day
daily_message_count = messages.groupby('DATE_ONLY').size()

# Line graph for message volume over time
daily_message_count.plot(kind='line', figsize=(10, 6), color='blue')
plt.title('Message Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Messages')
plt.grid(True)
plt.show()
```

```python
# Calculate the length of each message
messages['CONTENT_LENGTH'] = messages['CONTENT'].apply(lambda x: len(str(x)))

# Histogram for the distribution of message lengths
messages['CONTENT_LENGTH'].plot(kind='hist', bins=20, color='purple', edgecolor='black')
plt.title('Distribution of Message Lengths')
plt.xlabel('Message Length')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Extract the hour from the time
messages['HOUR'] = messages['DATE'].dt.hour

# Group by hour to count the number of messages
hourly_message_count = messages.groupby('HOUR').size()

# Bar graph for most active hours
hourly_message_count.plot(kind='bar', color='orange')
plt.title('Most Active Hours of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Messages')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
from textblob import TextBlob

# Function to calculate sentiment polarity
def get_sentiment(text):
    return TextBlob(str(text)).sentiment.polarity

# Apply sentiment analysis to the content
messages['SENTIMENT'] = messages['CONTENT'].apply(get_sentiment)

# Histogram for sentiment distribution
messages['SENTIMENT'].plot(kind='hist', bins=20, color='teal', edgecolor='black')
plt.title('Distribution of Sentiment Polarity')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all messages into a single string
all_words = ' '.join(str(text) for text in messages['CONTENT'])

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_words)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Top Words in Messages')
plt.axis('off')
plt.show()
```

```python

# Group by conversation ID to find the start and end times of each conversation
conversation_durations = messages.groupby('CONVERSATION ID')['DATE'].agg(['min', 'max'])

# Calculate the duration of each conversation
conversation_durations['DURATION'] = conversation_durations['max'] - conversation_durations['min']

# Plot the distribution of conversation durations
conversation_durations['DURATION'].dt.total_seconds().plot(kind='hist', bins=20, color='navy', edgecolor='black')
plt.title('Distribution of Conversation Durations')
plt.xlabel('Duration (seconds)')
plt.ylabel('Number of Conversations')
plt.grid(True)
plt.show()

```

```python
# Sort messages by conversation ID and date
messages_sorted = messages.sort_values(by=['CONVERSATION ID', 'DATE'])

# Calculate the time difference between consecutive messages in the same conversation
messages_sorted['RESPONSE_TIME'] = messages_sorted.groupby('CONVERSATION ID')['DATE'].diff()

# Filter out NaT values (first message in each conversation)
response_times = messages_sorted['RESPONSE_TIME'].dropna()

# Plot the distribution of response times
response_times.dt.total_seconds().plot(kind='hist', bins=50, color='green', edgecolor='black')
plt.title('Distribution of Response Times')
plt.xlabel('Response Time (seconds)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Extract day of the week and hour of the day
messages['DAY_OF_WEEK'] = messages['DATE'].dt.day_name()
messages['HOUR'] = messages['DATE'].dt.hour

# Create a pivot table to count messages by day and hour
heatmap_data = messages.pivot_table(index='DAY_OF_WEEK', columns='HOUR', aggfunc='size', fill_value=0)

# Reorder the days of the week
heatmap_data = heatmap_data.reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Plot the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='Blues', annot=True, fmt='d')
plt.title('Heatmap of Communication Activity by Day and Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Day of the Week')
plt.show()
```

```python
import networkx as nx

# Create an empty graph
G = nx.Graph()

# Add edges between senders and recipients
for index, row in messages.iterrows():
    G.add_edge(row['FROM'], row['TO'])

# Plot the network graph
plt.figure(figsize=(30, 30))
nx.draw_networkx(G, with_labels=True, node_color='skyblue', edge_color='gray', node_size=3000, font_size=10, font_color='black')
plt.title('Communication Network')
plt.show()
```

```python
# Apply sentiment analysis (using TextBlob or another library)
messages['SENTIMENT'] = messages['CONTENT'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Group by date and calculate average sentiment per day
sentiment_over_time = messages.groupby('DATE_ONLY')['SENTIMENT'].mean()

# Plot sentiment over time
sentiment_over_time.plot(kind='line', color='magenta', figsize=(10, 6))
plt.title('Average Sentiment Over Time')
plt.xlabel('Date')
plt.ylabel('Average Sentiment')
plt.grid(True)
plt.show()
```

```python
# Calculate the correlation between message length and response time
correlation = messages_sorted[['CONTENT_LENGTH', 'RESPONSE_TIME']].corr().iloc[0, 1]

# Scatter plot of message length vs. response time
plt.figure(figsize=(8, 6))
plt.scatter(messages_sorted['CONTENT_LENGTH'], messages_sorted['RESPONSE_TIME'].dt.total_seconds(), alpha=0.5, color='orange')
plt.title(f'Correlation Between Message Length and Response Time: {correlation:.2f}')
plt.xlabel('Message Length')
plt.ylabel('Response Time (seconds)')
plt.grid(True)
plt.show()
```

```python
# Create a new column for the week number
messages['WEEK'] = messages['DATE'].dt.isocalendar().week

# Group by week to count the number of messages per week
weekly_message_count = messages.groupby('WEEK').size()

# Line graph for weekly communication trends
weekly_message_count.plot(kind='line', color='darkblue', figsize=(10, 6))
plt.title('Weekly Communication Trends')
plt.xlabel('Week Number')
plt.ylabel('Number of Messages')
plt.grid(True)
plt.show()
```

```python
# Filter messages for the top 5 senders
top_5_senders = top_senders.index[:5]
filtered_messages = messages[messages['FROM'].isin(top_5_senders)]

# Group by sender and date to count messages per day
sender_activity = filtered_messages.groupby(['FROM', 'DATE_ONLY']).size().unstack(fill_value=0)

# Plot the activity of each sender over time
sender_activity.T.plot(kind='line', figsize=(12, 6))
plt.title('Activity of Top 5 Senders Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Messages')
plt.legend(title='Sender')
plt.grid(True)
plt.show()
```

```python
# Stacked bar chart for top sender-recipient pairs
sender_recipient = messages.pivot_table(index='FROM', columns='TO', aggfunc='size', fill_value=0)
top_sender_recipient = sender_recipient.loc[top_senders.index[:5]]

top_sender_recipient.plot(kind='bar', stacked=True, figsize=(20,10), colormap='viridis')
plt.title('Top Senders and Their Recipients')
plt.xlabel('Sender')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns

# Violin plot for response times by sender
sns.violinplot(x='FROM', y='RESPONSE_TIME', data=messages_sorted, inner='quartile', palette='muted')
plt.title('Distribution of Response Times by Sender')
plt.xlabel('Sender')
plt.ylabel('Response Time (seconds)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot for message length vs. time of day
plt.scatter(messages['TIME_ONLY'].apply(lambda x: x.hour + x.minute/60), messages['CONTENT_LENGTH'], alpha=0.5)
plt.title('Message Length vs. Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Message Length')
plt.grid(True)
plt.show()
```

```python
from math import pi

# Define data for radar chart
categories = ['Message Length', 'Frequency', 'Response Time']
num_vars = len(categories)
angles = [n / float(num_vars) * 2 * pi for n in range(num_vars)]
angles += angles[:1]

# Example data for two senders
sender_data = {
    'Sender 1': [120, 50, 30],
    'Sender 2': [80, 70, 45]
}

# Plotting the radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

# Draw one line per sender
for sender, values in sender_data.items():
    values += values[:1]
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=sender)
    ax.fill(angles, values, alpha=0.25)

ax.set_xticks(angles[:-1])
ax.set_xticklabels(categories)
plt.title('Sender Comparison Across Multiple Metrics')
plt.legend()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# 3D scatter plot for message length, response time, and time of day
ax.scatter(messages_sorted['TIME_ONLY'].apply(lambda x: x.hour + x.minute/60),
           messages_sorted['CONTENT_LENGTH'],
           messages_sorted['RESPONSE_TIME'].dt.total_seconds(), alpha=0.5)

ax.set_xlabel('Time of Day')
ax.set_ylabel('Message Length')
ax.set_zlabel('Response Time (seconds)')
plt.title('3D Plot of Message Characteristics')
plt.show()
```
